```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
url = 'https://www.tinkoff.ru/invest/social/profile/Vlad_pro_Dengi/'
def func(url):

    driver = webdriver.Chrome()

    driver.get(url)

    try:
        for i in range(5):  
            print('start', i)
            time.sleep(100)
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight - 1000);")
            # WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.TAG_NAME, 'footer')))
            time.sleep(1)

        page_source = driver.page_source
        
        soup = BeautifulSoup(page_source, 'html.parser')

        posts = soup.find_all('div', {'data-qa-tag':'PulsePost'})
        l=[]
        for post in posts:
            l.append(post.find_all('div', {'class':'pulse-profilepage__fGGBmY'})[0].get_text())
        print(len(l))
        return l
    finally:
        # Close the WebDriver when done
        driver.quit()
l = func(url)
print(l[0])
```
