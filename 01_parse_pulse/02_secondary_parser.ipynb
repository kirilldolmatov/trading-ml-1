```python
import requests

# Replace these values with your actual API endpoint and parameters
api_url = 'https://api.tinkoff.com/pulse/v1/data'
api_key = 'your_api_key'
author_name = 'Tdmap'

# Set up the request headers with the API key
headers = {
    'Authorization': f'Bearer {api_key}',
}

# Set up the request parameters, including the author parameter
params = {
    'author': author_name,
}

# Make the API request
response = requests.get(api_url, headers=headers, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse and work with the JSON response
    data = response.json()
    print(data)
else:
    # Print an error message if the request was not successful
    print(f"Error: {response.status_code} - {response.text}")
```
